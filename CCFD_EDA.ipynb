```python
credit_cards_nofraud = credit_cards[credit_cards['Class']==0]
credit_cards_fraud = credit_cards[credit_cards['Class']==1]
```

```python
def distplots(columns, ncol=None, figsize=(30, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        #ax.scatter(x = col, y = 'Class', data=credit_cards, alpha=0.5)
        sns.distplot(credit_cards_nofraud[col],ax=ax)
        ax.set(xlabel=col, ylabel='')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
def distplots_fraud(columns, ncol=None, figsize=(30, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        #ax.scatter(x = col, y = 'Class', data=credit_cards, alpha=0.5)
        sns.distplot(credit_cards_fraud[col],ax=ax)
        ax.set(xlabel=col, ylabel='')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
credit_cards.Class.unique()
```

```python
features = [x for x in credit_cards.columns if x not in ['Class']]
```

```python
distplots(features, ncol=4, figsize=(30, 30))
distplots_fraud(features, ncol=4, figsize=(30, 30))
```
